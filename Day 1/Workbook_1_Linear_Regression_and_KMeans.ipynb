## Section D: KMeans -- Fill in the Blanks

Complete the KMeans assignment step.

```python
import numpy as np
from sklearn.datasets import make_blobs

X_data, _ = make_blobs(n_samples=100, centers=3, random_state=42)

k = 3
np.random.seed(0)

# Step 1: Initialize centroids by picking k random data points
indices = np.random.choice(len(X_data), k, replace=False)
centroids = X_data[indices].copy()

for iteration in range(20):
    # Step 2: Assign each point to the nearest centroid
    distances = np.linalg.norm(
        X_data[:, np.newaxis] - _____,  # FILL IN: what are we measuring distance to?
        axis=2
    )
    labels = np._____(distances, axis=1)  # FILL IN: which numpy function gives the index of the minimum?

    # Step 3: Update centroids to the mean of assigned points
    new_centroids = np.array([
        X_data[labels == i]._____(axis=0)  # FILL IN: what operation computes the center?
        for i in range(k)
    ])

    centroids = new_centroids

print(f'Final centroids:\n{centroids}')
print(f'\nCluster sizes: {[np.sum(labels == i) for i in range(k)]}')
```
